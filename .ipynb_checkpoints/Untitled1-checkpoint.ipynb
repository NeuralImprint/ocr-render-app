```python
from paddleocr import PaddleOCR, draw_ocr
from PIL import Image
import matplotlib.pyplot as plt

# Initialize the OCR model
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Ensure models are downloaded

# Path to your image
img_path = 'Excel.png','Hand.jpg'  # Replace with your image path

# Run OCR
results = ocr.ocr(img_path, cls=True)

# Visualize results
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in results[0]]
txts = [line[1][0] for line in results[0]]
scores = [line[1][1] for line in results[0]]

# Draw result
annotated = draw_ocr(image, boxes, txts, scores, font_path='C:/Windows/Fonts/arial.ttf')
plt.figure(figsize=(15, 10))
plt.imshow(annotated)
plt.axis('off')
plt.show()
```

```python
from paddleocr import PPStructure, save_structure_res
import pandas as pd
import os

# Step 1: Prepare directories and image path
os.makedirs('./output', exist_ok=True)
image_path = 'Excel.png'  # Replace with your image

# Step 2: Initialize table structure OCR
ocr_engine = PPStructure(
    layout=True,
    show_log=True,
    table=True,
    ocr=True,
    lang='en',
    structure_version='PP-StructureV2'
)

# Step 3: Run structure OCR
result = ocr_engine(image_path)

# Step 4: Save the output structure
save_structure_res(result, './output', os.path.basename(image_path))


# Step 5: Convert to DataFrame
for res in result:
    if 'res' in res and isinstance(res['res'], list):
        table_data = res['res']
        rows = []
        for row in table_data:
            rows.append([cell['text'] for cell in row])

        # Convert to pandas table
        df = pd.DataFrame(rows[1:], columns=rows[0])  # First row = header
        print(df)

        # Optional: Save as CSV
        df.to_csv('ocr_table.csv', index=False)
```

```python
from paddleocr import PaddleOCR
from PIL import Image
import pandas as pd
import os

# Initialize OCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Use your config

# List of image paths
image_paths = ['Hand.jpg', 'Excel.png']

# Output folder
output_folder = './output'
os.makedirs(output_folder, exist_ok=True)

# Process each image
for image_path in image_paths:
    print(f"Processing: {image_path}")
    
    # Run OCR
    result = ocr.ocr(image_path, cls=True)
    
    # Prepare data for table
    data = []
    for line in result[0]:
        text = line[1][0]
        confidence = line[1][1]
        data.append({'Text': text, 'Confidence': confidence})
    
    # Convert to DataFrame
    df = pd.DataFrame(data)

    # Save each result as a separate file
    filename = os.path.splitext(os.path.basename(image_path))[0]
    df.to_excel(os.path.join(output_folder, f"{filename}_ocr_output.xlsx"), index=False)

    print(f"Saved: {filename}_ocr_output.xlsx")

print("✅ All images processed and saved.")
```

```python
from paddleocr import PPStructure, save_structure_res
import pandas as pd
import os

# Step 1: Create output directory
os.makedirs('./output', exist_ok=True)
image_path = 'Onlhand.png'  # Your image path

# Step 2: Initialize structure OCR
ocr_engine = PPStructure(
    layout=True,
    show_log=True,
    table=True,
    ocr=True,
    lang='en',
    structure_version='PP-StructureV2'
)

# Step 3: Run OCR
result = ocr_engine(image_path)

# Step 4: Save structure result
save_structure_res(result, './output', os.path.basename(image_path))

# Step 5: Extract table text safely
all_rows = []

for res in result:
    if 'res' in res and isinstance(res['res'], list):
        for row in res['res']:
            if isinstance(row, list):
                # Handle case where cells are dicts with 'text'
                if all(isinstance(cell, dict) and 'text' in cell for cell in row):
                    all_rows.append([cell['text'] for cell in row])
                # Handle case where cells are already plain strings
                elif all(isinstance(cell, str) for cell in row):
                    all_rows.append(row)

# Step 6: Always return a DataFrame — even if empty
try:
    df = pd.DataFrame(all_rows[1:], columns=all_rows[0]) if all_rows else pd.DataFrame()
except Exception:
    df = pd.DataFrame(all_rows)  # fallback to generic table

# Step 7: Display and save
print(df)
df.to_csv("ocr_table.csv", index=False)
```

```python
from paddleocr import PaddleOCR
import pandas as pd

ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Use normal OCR pipeline
image_path = 'Onlhand.png'
result = ocr.ocr(image_path, cls=True)

# Flatten results into rows of text
rows = []
for line in result[0]:
    text = line[1][0]
    rows.append([text])

# Create DataFrame
df = pd.DataFrame(rows, columns=["Text"])
print(df)
df.to_csv("ocr_text_output.csv", index=False)
```

```python
from paddleocr import PaddleOCR
import pandas as pd

ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Initialize OCR
image_path = 'Onlhand.png'
result = ocr.ocr(image_path, cls=True)

# Extract all text strings into a list
texts = [line[1][0] for line in result[0]]

# Split into header and rows
num_cols = 4  # Since you have Name, Age, Marks, Roll No
header = texts[:num_cols]
data_rows = [texts[i:i+num_cols] for i in range(num_cols, len(texts), num_cols)]

# Convert to DataFrame
df = pd.DataFrame(data_rows, columns=header)
print(df)

# Optional: Save as CSV
df.to_csv("structured_ocr_table.csv", index=False)
```
